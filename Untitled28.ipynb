```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate example time series data
np.random.seed(0)
dates = pd.date_range('2023-01-01', periods=365)
values = np.random.randn(365).cumsum()

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Value': values})

# Plot time series data with Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Value', data=df)
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()

# Generate example of a related variable (e.g., temperature)
temperature = np.random.randint(0, 30, 365)

# Add temperature data to the DataFrame
df['Temperature'] = temperature

# Plot relationship between time series data and temperature using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Value', y='Temperature', data=df)
plt.title('Relationship between Time Series Data and Temperature')
plt.xlabel('Value')
plt.ylabel('Temperature')
plt.show()
```
